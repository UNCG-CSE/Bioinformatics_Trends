```python
import csv
import pandas as pd
from ast import literal_eval

df_logan = pd.read_csv(r'/home/logan/Documents/DataScience/csv/Merged_all_years/all_years_merged(logan,mouna,darpan).csv',
                       encoding='cp1252' ,dtype='str', error_bad_lines=False)

# get dimmensions of dataframe to ensure csv imported correctly
df_logan.shape
```

```python
df_thCSC = df_logan.loc[df_logan['Text'].isin(['Theoretical Computer Science'])]
df_genCSC = df_logan.loc[df_logan['Text'].isin(['Computer Science (all)'])]
df_molBIO = df_logan.loc[df_logan['Text'].isin(['Molecular Biology'])]
df_GEN = df_logan.loc[df_logan['Text'].isin(['Genetics'])]
df_appCSC = df_logan.loc[df_logan['Text'].isin(['Computer Science Applications'])]
df_BIOcheMB = df_logan.loc[df_logan['Text'].isin(['Biochemistry, Genetics and Molecular Biology (all)'])]
df_BIOMedEng = df_logan.loc[df_logan['Text'].isin(['Biomedical Engineering'])]
df_BIOchem = df_logan.loc[df_logan['Text'].isin(['Biochemistry'])]
df_MED = df_logan.loc[df_logan['Text'].isin(['Medicine (all)'])]
```

```python
df_thCSC['cite'] = pd.to_numeric(df_thCSC['cite'], downcast=None)
df_genCSC['cite'] = pd.to_numeric(df_genCSC['cite'], downcast=None)
df_molBIO['cite'] = pd.to_numeric(df_molBIO['cite'], downcast=None)
df_GEN['cite'] = pd.to_numeric(df_GEN['cite'], downcast=None)
df_appCSC['cite'] = pd.to_numeric(df_appCSC['cite'], downcast=None)
df_BIOcheMB['cite'] = pd.to_numeric(df_BIOcheMB['cite'], downcast=None)
df_BIOMedEng['cite'] = pd.to_numeric(df_BIOMedEng['cite'], downcast=None)
df_BIOchem['cite'] = pd.to_numeric(df_BIOchem['cite'], downcast=None)
df_MED['cite'] = pd.to_numeric(df_MED['cite'], downcast=None)
```
