Read merged (all years) CSV file into data frame:

```python
import csv
import pandas as pd
from ast import literal_eval

df_logan = pd.read_csv(r'/Users/loganhornbuckle/Desktop/Logan/Merged/all_years_merged_logan_mouna_darpan.csv',
                       encoding='cp1252' ,dtype='str', error_bad_lines=False)

# get dimmensions of dataframe to ensure csv imported correctly
df_logan.shape
```

```python
df_logan['cite'] = pd.to_numeric(df_logan['cite'], downcast=None)
```

```python
df_thCSC = df_logan.loc[df_logan['Text'].isin(['Theoretical Computer Science'])]
df_genCSC = df_logan.loc[df_logan['Text'].isin(['Computer Science (all)'])]
df_molBIO = df_logan.loc[df_logan['Text'].isin(['Molecular Biology'])]
df_GEN = df_logan.loc[df_logan['Text'].isin(['Genetics'])]
df_appCSC = df_logan.loc[df_logan['Text'].isin(['Computer Science Applications'])]
df_BIOcheMB = df_logan.loc[df_logan['Text'].isin(['Biochemistry, Genetics and Molecular Biology (all)'])]
df_BIOMedEng = df_logan.loc[df_logan['Text'].isin(['Biomedical Engineering'])]
df_BIOchem = df_logan.loc[df_logan['Text'].isin(['Biochemistry'])]
df_MED = df_logan.loc[df_logan['Text'].isin(['Medicine (all)'])]
df_MATH = df_logan.loc[df_logan['Text'].isin(['Applied Mathematics'])]
df_BIOtech = df_logan.loc[df_logan['Text'].isin(['Biotechnology'])]
```

# Compare distributions of Most cited/published subjects

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(25,20)})

sns.distplot(df_appCSC['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'App CSC')
sns.distplot(df_GEN['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Genetics')
sns.distplot(df_BIOcheMB['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'BioChem')
sns.distplot(df_MATH['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Math')
sns.distplot(df_molBIO['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Mol Bio')

plt.xlim(0,200)
```
