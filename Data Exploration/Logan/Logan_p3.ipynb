# Breakdown of avg citations per paper (grouped by field of study) for the following:
- top 10 most published
- top 10 most cited
- top 2 most published
- top 2 most cited

Read merged (all years) CSV file into data frame:

```python
import csv
import pandas as pd
from ast import literal_eval

df_logan = pd.read_csv(r'/Users/loganhornbuckle/Desktop/Logan/Merged/all_years_merged_logan_mouna_darpan.csv',
                       encoding='cp1252' ,dtype='str', error_bad_lines=False)

# get dimmensions of dataframe to ensure csv imported correctly
df_logan.shape
```

```python
df_logan.describe()
```

```python
del df_logan['Author Keyword']
del df_logan['eid']
del df_logan['Lang']
del df_logan['Para']
del df_logan['Title']
del df_logan['year']
del df_logan['publication_name']
del df_logan['Code']
del df_logan['Abbreviation']
del df_logan['publication_type']
```

```python
df_logan
```

```python
# Sum number of publications for each field

df_logan['cite'] = pd.to_numeric(df_logan['cite'])
df_count = df_logan.groupby('Text').sum()
df_count['Num_Publications'] = df_logan['Text'].value_counts()
```

```python
df_count
```

```python
# create new column: avg citations per publication

df_count['Cit_per_Pub'] = df_count['cite']/df_count['Num_Publications']
df_stat = df_count.sort_values(by = 'Cit_per_Pub', ascending=False)
```

```python
df_stat
```

```python
df_stat.reset_index()
```

```python
# take only top 10 fields (number publications)

df_stat = df_stat.sort_values(by = 'Num_Publications', ascending=False)
df_top10Pub = df_stat.head(10)
df_top10Pub
```

```python
df_top10Pub = df_top10Pub.sort_values(by = 'Cit_per_Pub', ascending=False)
```

```python
del df_top10Pub['cite']
del df_top10Pub['Num_Publications']
```

```python
# avg number citations -- top 10 most published fields

ax = df_top10Pub.plot.bar(rot=90, fontsize=20 , figsize=(20,15))
```

```python
df_top10Pub['Cit_per_Pub'] = pd.to_numeric(df_top10Pub['Cit_per_Pub'], downcast=None)
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(20,15)})
sns.set(font_scale=3)

sns.kdeplot(df_top10Pub['Cit_per_Pub'],   bw=0.2);
```

# Same approach, but for most cited instead of most published

```python
df_stat = df_stat.sort_values(by = 'cite', ascending=False)
df_stat
```

```python
df_top10Cite = df_stat.head(10)
df_top10Cite
```

```python
del df_top10Cite['Num_Publications']
del df_top10Cite['cite']
```

```python
# avg number citations -- top 10 most cited fields
df_top10Cite = df_top10Cite.sort_values(by = 'Cit_per_Pub', ascending=False)

df_top10Cite
```

```python
ax = df_top10Cite.plot.bar(rot=90 ,fontsize=20, figsize=(15,12))
ax.legend(["avg number citations -- top 10 most cited fields"]);
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.set(font_scale=3)

sns.kdeplot(df_top10Cite['Cit_per_Pub'], bw=0.2);
```
