```python
import pandas as pd

df_darpan = pd.read_csv(r'C:\Users\Darpan\Desktop\Subjects\Data Science\Merged data (mouna,logan,darpan)\All Years Merged File\all_years_merged(logan,mouna,darpan).csv',encoding='cp1252', dtype='str', error_bad_lines=False)
df_darpan.shape
```

```python
df_darpan[:10]
```

```python
del df_darpan['Lang']
del df_darpan['Para']
del df_darpan['year']
del df_darpan['publication_name']
del df_darpan['publication_type']
del df_darpan['Author Keyword']
del df_darpan['Abbreviation']
del df_darpan['Code']
del df_darpan['Text']
df_darpan[:5]
```

```python
df_darpan.dropna(inplace=True)
df_darpan.shape
```

```python
df_darpan['Title'] = df_darpan['Title'].str.split().str.len()
df_darpan[:5]
```

```python
df_darpan['cite'] = pd.to_numeric(df_darpan['cite'])
```

```python
df_darpan['cite'].sort_values()
```

```python
df_darpan[['cite','Title']].corr()
```

```python
df_darpan = df_darpan[df_darpan.cite < 9000]
df_darpan = df_darpan[df_darpan.Title < 48]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(df_darpan['Title'], df_darpan['cite'])
plt.xlabel('Length of Title')
plt.ylabel('Number of Citations')
plt.show()
```
