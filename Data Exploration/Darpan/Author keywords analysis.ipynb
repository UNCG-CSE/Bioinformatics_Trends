```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Darpan\Desktop\Subjects\Data Science\Merged data (mouna,logan,darpan)\All Years Merged File\all_years_merged(logan,mouna,darpan).csv',encoding='cp1252', dtype='str', error_bad_lines=False)
df[:10]
```

```python
df.shape
```

```python
df = df.dropna(subset=['Author_Keyword'])
df.shape
```

```python
df[:10]
```

```python
df.drop(['eid','Lang', 'Para', 'Title', 'year', 'publication_name', 'publication_type', 'Abbreviation', 'Code', 'Text'], axis=1, inplace=True)
df[:10]
```

```python
df['Author_Keyword'] = df['Author_Keyword'].map(lambda x: x.lstrip('[').rstrip(']'))
df[:10]
```

```python
df = df.set_index('cite').Author_Keyword.str.split(',', expand=True).stack().reset_index('cite')
df.columns = ['cite', 'Author_Keyword']
df.reset_index(drop=True, inplace=True)
df[:10]
```

```python
df.shape
```

```python
df['Author_Keyword'] = df['Author_Keyword'].str.replace("'", "")
df[:10]
```

```python
#Removing white space before and after each word
df['Author_Keyword'] = df['Author_Keyword'].str.strip()
df[:10]
```

```python
df1=df.copy()
df1[:5]
```

```python
df1['Counts'] = df1.groupby(['Author_Keyword'])['cite'].transform('count')
df1[:50]
```

```python
df1.loc[df1['Author_Keyword'] == "Reducts"]
```

```python
df2=df1.copy()
```

```python
df2.sort_values("Counts", ascending=False, inplace=True)
df2[:10]
```

```python
df2["cite"] = pd.to_numeric(df2["cite"])
df2 = df2.groupby(['Author_Keyword','Counts'], as_index=False)['cite'].mean()
df2.sort_values("Counts", ascending=False, inplace=True)
df2[:11]
```

```python
df2 = df2.head(11)
df2 = df2.drop(df2.index[[2]])
df2
```

```python
df3=df1.copy()
df3["cite"] = pd.to_numeric(df3["cite"])
#Use only keywords which have appeared in more than 100 papers
df3 = df3[df3.Counts > 100]
df3 = df3.groupby(['Author_Keyword','Counts'], as_index=False)['cite'].mean()
df3.sort_values("cite", ascending=False, inplace=True)
df3[:50]
```

```python
df["cite"] = pd.to_numeric(df["cite"])
df_Bioinformatics = df.loc[df['Author_Keyword'].isin(['Bioinformatics'])]
df_Proteomics = df.loc[df['Author_Keyword'].isin(['Proteomics'])]
df_Genomics = df.loc[df['Author_Keyword'].isin(['Genomics'])]
df_Gene_expression = df.loc[df['Author_Keyword'].isin(['Gene expression'])]
df_Microarray = df.loc[df['Author_Keyword'].isin(['Microarray'])]


df_Support_vector_machines = df.loc[df['Author_Keyword'].isin(['Support vector machines'])]
df_Neural_networks = df.loc[df['Author_Keyword'].isin(['Neural networks'])]
df_Inflammation = df.loc[df['Author_Keyword'].isin(['Inflammation'])]
df_Functional_genomics = df.loc[df['Author_Keyword'].isin(['Functional genomics'])]
df_Mass_spectrometry = df.loc[df['Author_Keyword'].isin(['Mass spectrometry'])]
```

```python
#Top 5 based on frequency

import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(df_Bioinformatics['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Bioinformatics')
sns.distplot(df_Proteomics['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Proteomics')
sns.distplot(df_Genomics['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Genomics')
sns.distplot(df_Gene_expression['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Gene expression')
sns.distplot(df_Microarray['cite'], hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 3}, label = 'Microarray')

plt.ylim(0,0.00060)
plt.xlim(0,1000)
plt.show()
```

```python
#Top 5 based on frequency

df_top5frequency = df_Bioinformatics.append(df_Proteomics, ignore_index=True)
df_top5frequency = df_top5frequency.append(df_Genomics, ignore_index=True)
df_top5frequency = df_top5frequency.append(df_Gene_expression, ignore_index=True)
df_top5frequency = df_top5frequency.append(df_Microarray, ignore_index=True)

sns.boxplot(y='Author_Keyword', x='cite', data=df_top5frequency)
plt.xlim(0, 400)
plt.show()
```
