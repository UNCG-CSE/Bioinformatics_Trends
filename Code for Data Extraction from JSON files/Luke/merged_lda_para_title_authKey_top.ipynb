```python
# cleaning data
test = test.str.replace(']', ' ')
test = test.str.replace('[', ' ')
test = test.str.replace('(', ' ')
test = test.str.replace(')', ' ')
test = test.str.replace(',', ' ')
test = test.str.replace('  ', ' ')
test = test.str.replace('  ', ' ')

# cleaning the data further...

# for every paper
for x in range (0, len(test)):

    # create list of every element
    test[x] = list(map(str, test[x].split(' ')))
    
    #remove leftover empty entries
    for y in range (0, len(test[x])-1):
        if test[x][y] == '':
            del test[x][y]
    
    # change everything from string to float
    test[x] = list(map(float, test[x]))
```

```python
test2 = test.copy()
```

```python
print(len(test))
print(len(test2))
```

```python
# saving number of relevant topics for this column

lda_probMat_para_numTopics = [0] * 79630

for paper in range (0, len(test2)):
    lda_probMat_para_numTopics[paper] = int(len(test2[paper]) / 2)
```

```python
print(lda_probMat_para_numTopics[0:5])
print(test2[2])
```

```python
# finding max 

# step through every paper
for paper in range (0, len(test2)):
    
    # for storing the highest associated topic
    maxIndi = 1
    maxNum = 0.0
    
    # step through every topic for this paper
    for corr in range (0, len(test2[paper])):
        # if its a correlation
        if test2[paper][corr] < 1.0:
            # if this correlation is higher than the others that have come before it
            if test2[paper][corr] > test2[paper][maxIndi]:
                # this is the new high for this paper
                maxIndi = corr
    # now 'maxIndi' contains the highest correlation
    test2[paper] = [test2[paper][maxIndi-1], test2[paper][maxIndi]]
```
